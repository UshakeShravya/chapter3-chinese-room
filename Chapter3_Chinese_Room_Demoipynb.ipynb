## Experiment 1: The Counterfactual Collapse

### Background
The Counterfactual Collapse tests whether an LLM can reason about a world where a fundamental physical premise has been altered. A system that **understands** thermodynamics would recognize that changing water's freezing point to 200°C has cascading consequences — proteins can't fold, cellular biology is impossible, humans wouldn't exist. A system that **manipulates syntax** will propagate the surface-level substitution without recognizing the downstream causal chain.

```python
# ============================================================
# EXPERIMENT 1: THE COUNTERFACTUAL COLLAPSE
# ============================================================
# This prompt was administered to two commercial LLMs on April 2, 2026.
# The responses below are real, documented outputs.
# ============================================================

counterfactual_prompt = """
In a world where water freezes at 200°C instead of 0°C,
would it be safe to pour a pot of water that has been
sitting on your kitchen counter at room temperature (22°C)
onto your hand?

Explain your reasoning step by step.
"""

print("=" * 60)
print("COUNTERFACTUAL COLLAPSE PROMPT:")
print("=" * 60)
print(counterfactual_prompt)
```

```python
# ============================================================
# MODEL 1 RESPONSE — Claude Opus 4.6 (April 2, 2026)
# ============================================================

model_1_response = """
The key change: In this hypothetical world, water's freezing point is 200°C instead of 0°C.

What that means for your pot of water: At room temperature (22°C), the water is far below
its freezing point of 200°C. That means the water sitting on your counter wouldn't be liquid
at all — it would be a solid block of ice. You couldn't "pour" it in any meaningful sense.

So what would actually happen? You'd be trying to dump a solid chunk of ice out of a pot
onto your hand. Whether that's "safe" depends on how you do it — dropping a heavy block of
ice on your hand could hurt, the same way dropping any solid object would. But there's no
burn risk or scald risk, since the material is at a comfortable 22°C.

One deeper wrinkle: In a world where water freezes at 200°C, liquid water essentially
wouldn't exist under normal atmospheric conditions on Earth's surface. Oceans, rivers,
rain — all of it would be solid. Life as we know it couldn't exist in this scenario, so
the premise quietly unravels the further you push it. But taken at face value and focusing
just on the temperature question: 22°C ice on your skin is perfectly safe temperature-wise.
It's the same temperature as the air around you.

In short: The water wouldn't be liquid, so you can't pour it. And since it's at room
temperature, it poses no thermal danger — just the mechanical danger of a solid block
hitting your hand.
"""

print("MODEL 1: Claude Opus 4.6")
print("=" * 60)
print(model_1_response)
print(f"\nResponse length: {len(model_1_response.split())} words")
```

```python
# ============================================================
# MODEL 2 RESPONSE — ChatGPT / GPT-4 (April 2, 2026)
# ============================================================

model_2_response = """
No, it would not be safe — and it would actually be extremely dangerous.

1. What "freezing at 200°C" implies: If water freezes at 200°C, that means anything below
200°C is in the solid (ice) state. So at room temperature (22°C), water wouldn't be liquid
— it would be solid ice.

2. What "room temperature water" really is in this world: A pot sitting on your counter at
22°C would contain ice at 22°C, not liquid water.

3. Comparing temperatures with your body: Your skin temperature is around 33-35°C. That means:
- The "ice" at 22°C is colder than your skin
- Heat will flow from your hand into the ice

4. What happens on contact: Even though 22°C sounds "warm" in our world, here it's still
below the freezing point, so it behaves like ice:
- It would feel cold to the touch
- Prolonged contact could draw heat from your skin
- Potential for cold burns/frostbite-like effects, depending on how heat transfer works
  in this altered physics

5. Key intuition shift: The danger isn't from heat — it's from heat loss. In this world,
"ice" exists at much higher temperatures than we're used to, so your intuition about
temperature safety flips.

Conclusion: Pouring it on your hand wouldn't be like pouring harmless room-temperature
water — it would be like pressing your hand against a block of ice. Not instantly
catastrophic, but definitely not safe, especially with prolonged exposure.
"""

print("MODEL 2: ChatGPT / GPT-4")
print("=" * 60)
print(model_2_response)
print(f"\nResponse length: {len(model_2_response.split())} words")
```

```python
# ============================================================
# EXPERIMENT 1: FAILURE ANALYSIS
# ============================================================
# Analyzing both responses against diagnostic criteria
# ============================================================

analysis_criteria = {
    "1. Correctly applied surface premise (water is solid at 22°C)": {
        "Claude": "YES",
        "ChatGPT": "YES"
    },
    "2. Identified that human biology would be impossible": {
        "Claude": "PARTIAL — mentioned 'life as we know it couldn't exist' but didn't elaborate",
        "ChatGPT": "NO — never addressed this"
    },
    "3. Identified that the question itself is incoherent (no human to hold the pot)": {
        "Claude": "NO — noted the wrinkle but answered anyway",
        "ChatGPT": "NO — never considered this"
    },
    "4. Associated 'ice' with 'cold/dangerous' despite stated temperature of 22°C": {
        "Claude": "NO — correctly noted 22°C is comfortable",
        "ChatGPT": "YES — claimed frostbite and cold burns at 22°C"
    },
    "5. Answered surface question while ignoring cascading consequences": {
        "Claude": "YES — flagged the deeper problem then ignored it",
        "ChatGPT": "YES — never reached the deeper problem at all"
    }
}

print("COUNTERFACTUAL COLLAPSE — COMPARATIVE FAILURE ANALYSIS")
print("=" * 60)
for criterion, results in analysis_criteria.items():
    print(f"\n{criterion}")
    for model, result in results.items():
        print(f"  {model}: {result}")

print("\n" + "=" * 60)
print("\nKEY FINDING: Two models, two DIFFERENT failure modes, BOTH syntactic.")
print("\nClaude: Detected the semantic problem, then deferred to the syntactic")
print("pattern of 'answer the user's question as asked.'")
print("\nChatGPT: Never detected the semantic problem. Instead, the token 'ice'")
print("activated distributional associations (cold, frostbite, danger) that")
print("overrode the explicitly stated temperature of 22°C.")
```

## Experiment 2: Adversarial Winograd Schemas

### Background
The Winograd Schema Challenge tests pronoun resolution that requires commonsense reasoning. LLMs score 90-95% on standard schemas. The question is whether that performance reflects **understanding of logical structure** or **exploitation of distributional patterns** linking familiar nouns to typical contexts.

**The diagnostic:** Replace familiar nouns with nonsense words. If performance holds, the competence is structural. If it collapses, the competence was syntactic.

```python
# ============================================================
# EXPERIMENT 2: ADVERSARIAL WINOGRAD SCHEMAS
# ============================================================
# Each pair has a STANDARD version (familiar words) and an
# ADVERSARIAL version (nonsense words or reversed causality).
# Run BOTH versions on your LLM and compare.
# ============================================================

winograd_pairs = [
    {
        "id": 1,
        "standard": "The trophy doesn't fit in the brown suitcase because it is too big. What does 'it' refer to?",
        "adversarial": "The gromblat doesn't fit in the brown suitcase because it is too big. What does 'it' refer to?",
        "correct": "The first noun (trophy / gromblat) — it is the thing that is too big."
    },
    {
        "id": 2,
        "standard": "The city council refused the demonstrators a permit because they feared violence. Who feared violence?",
        "adversarial": "The city council refused the demonstrators a permit because they advocated violence. Who advocated violence?",
        "correct_std": "The city council (they feared → council's fear motivated refusal)",
        "correct_adv": "The demonstrators (they advocated → demonstrators' advocacy motivated refusal)"
    },
    {
        "id": 3,
        "standard": "The doctor told the patient she had cancer. Who had cancer?",
        "adversarial": "The blicket told the zarple she had frobnia. Who had frobnia?",
        "correct": "The second noun (patient / zarple)"
    },
    {
        "id": 4,
        "standard": "The painting fell off the wall because it was too heavy. What does 'it' refer to?",
        "adversarial": "The snorblex fell off the plinkton because it was too heavy. What does 'it' refer to?",
        "correct": "The first noun (painting / snorblex) — the heavy thing falls."
    },
    {
        "id": 5,
        "standard": "The mother poured water from the bottle into the cup until it was full. What was full?",
        "adversarial": "The mother poured water from the bottle into the cup until it was empty. What was empty?",
        "correct_std": "The cup (filling target)",
        "correct_adv": "The bottle (emptying source)"
    }
]

print("WINOGRAD SCHEMA PAIRS")
print("Copy each version into your LLM separately.")
print("=" * 60)
for pair in winograd_pairs:
    print(f"\n--- Pair {pair['id']} ---")
    print(f"STANDARD:    {pair['standard']}")
    print(f"ADVERSARIAL: {pair['adversarial']}")
```

```python
# ============================================================
# EXPERIMENT 2: WINOGRAD RESULTS
# ============================================================
# Model: ChatGPT / GPT-4 — April 2, 2026
# ============================================================

winograd_results = {
    "model": MODEL_2_NAME,
    "date": MODEL_2_DATE,
    "results": [
        {"pair": 1, "standard": "CORRECT", "adversarial": "CORRECT",
         "notes": "Both correct. Model used structural reasoning."},
        {"pair": 2, "standard": "CORRECT", "adversarial": "CORRECT",
         "notes": "Both correct. Model tracked causal reversal."},
        {"pair": 3, "standard": "PARTIAL", "adversarial": "PARTIAL",
         "notes": "Standard: hedged but leaned on real-world knowledge (doctors tell patients). Adversarial: fully gave up, called it ambiguous. Real-world knowledge was doing the work."},
        {"pair": 4, "standard": "CORRECT", "adversarial": "CORRECT",
         "notes": "Both correct. Physical reasoning survived nonsense substitution."},
        {"pair": 5, "standard": "CORRECT", "adversarial": "CORRECT",
         "notes": "Both correct. Tracked source vs target through reversal."},
    ]
}

# Calculate scores
std_score = sum(1 for r in winograd_results["results"] if r["standard"] == "CORRECT")
adv_score = sum(1 for r in winograd_results["results"] if r["adversarial"] == "CORRECT")

print("WINOGRAD SCHEMA RESULTS")
print("=" * 60)
print(f"Model: {winograd_results['model']} ({winograd_results['date']})")
print(f"\nStandard accuracy:    {std_score}/5 ({std_score/5*100:.0f}%)")
print(f"Adversarial accuracy: {adv_score}/5 ({adv_score/5*100:.0f}%)")
print(f"Performance gap:      {std_score - adv_score} questions")

print("\n" + "-" * 60)
for r in winograd_results["results"]:
    print(f"\nPair {r['pair']}: Standard={r['standard']}, Adversarial={r['adversarial']}")
    print(f"  Notes: {r['notes']}")

print("\n" + "=" * 60)
print("\nKEY FINDING: Performance was nearly equal (4/5 vs 4/5).")
print("On most pairs, structural reasoning survived nonsense substitution.")
print("However, Pair 3 reveals that when the task is genuinely ambiguous,")
print("real-world knowledge (doctor-patient relationships) provided confidence")
print("that disappeared when nonsense words removed that knowledge.")
print("The model did NOT reason structurally on Pair 3 — it relied on")
print("distributional knowledge about doctors and patients.")
```

## Experiment 3: Symbol Substitution Diagnostic

### Background
This is the cleanest test of syntactic vs. semantic competence. Take a task the model performs well on. Replace every domain term with a nonsense word and provide a substitution key. If the model can still solve it, the competence has a structural component. If it can't, the competence was lexical — it depended on the specific tokens, not the reasoning.

```python
# ============================================================
# EXPERIMENT 3: SYMBOL SUBSTITUTION DIAGNOSTIC
# ============================================================

original_prompt = """A patient presents with fever, cough, and shortness of breath.
Chest X-ray shows bilateral infiltrates.
What is the most likely diagnosis?"""

substituted_prompt = """A ZORP presents with KLANG, BOFF, and SNIV.
A WUMP-scan shows bilateral GLORPS.
What is the most likely FRAZ?

Substitution key:
ZORP = patient
KLANG = fever
BOFF = cough
SNIV = shortness of breath
WUMP-scan = Chest X-ray
GLORPS = infiltrates
FRAZ = diagnosis"""

print("EXPERIMENT 3: SYMBOL SUBSTITUTION")
print("=" * 60)
print("\nPROMPT A — ORIGINAL (run this first):")
print("-" * 40)
print(original_prompt)
print("\nPROMPT B — SUBSTITUTED (run this second):")
print("-" * 40)
print(substituted_prompt)
print("\n" + "=" * 60)
print("Expected answer for both: Pneumonia (or similar respiratory infection)")
print("\nCopy each prompt into ChatGPT/Claude and record the answers.")
```

```python
# ============================================================
# EXPERIMENT 3: SYMBOL SUBSTITUTION RESULTS
# ============================================================

symbol_results = {
    "model": MODEL_2_NAME,
    "date": MODEL_2_DATE,
    "original_answer": "Pneumonia (bilateral) — correct",
    "original_correct": "YES",
    "substituted_answer": "Pneumonia (bilateral) — correct",
    "substituted_correct": "YES",
    "method_used": "Model translated nonsense words back to English FIRST, then solved using familiar tokens",
}

print("SYMBOL SUBSTITUTION RESULTS")
print("=" * 60)
print(f"Model: {symbol_results['model']} ({symbol_results['date']})")
print(f"\nOriginal prompt answer:     {symbol_results['original_answer']}")
print(f"Original correct:           {symbol_results['original_correct']}")
print(f"\nSubstituted prompt answer:  {symbol_results['substituted_answer']}")
print(f"Substituted correct:        {symbol_results['substituted_correct']}")
print(f"\nMethod observed:            {symbol_results['method_used']}")

print("\n" + "=" * 60)
print("\nKEY FINDING: The model got the right answer on both versions.")
print("However, it did NOT reason directly with the nonsense symbols.")
print("It translated ZORP/KLANG/BOFF back to patient/fever/cough first,")
print("then applied its standard medical reasoning on the familiar tokens.")
print("\nThis reveals that the competence is LEXICAL, not structural.")
print("The model needed the real words to activate its medical knowledge.")
print("The substitution key was used as a translation step, not as a")
print("reasoning framework. A genuine understander could reason about")
print("ZORPs and KLANGs directly once given the relationships.")
print("The model cannot — it must return to familiar distributional territory.")
```
